```python
from metrixai.saliency_map import SaliencyMap
import pickle
import cv2

with open('image_person_0.pkl', 'rb') as f:
    x = pickle.load(f)

img = cv2.imread('image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
saliency_map = SaliencyMap(x, img, 0)
saliency_map.compute_insertion_curve(verbose=True)
```

```python
auc_insertion = saliency_map.compute_insertion_auc(verbose=True, divisions=10)
```

```python
auc_deletion = saliency_map.compute_deletion_auc(verbose=True, divisions=10)
```
